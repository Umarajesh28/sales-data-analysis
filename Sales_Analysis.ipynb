# Sales Data Analysis using Python (Pandas) & Visualization

***1. Create a Dataset***

```python


import pandas as pd

data = {
    "Product": ["A", "B", "C", "D", "E", "A", "B", "C", "D", "E"],
    "Sales": [100, 200, 150, 300, 250, 120, 220, 130, 310, 270],
    "Quantity": [2, 5, 3, 6, 4, 3, 6, 2, 7, 5],
    "Region": ["North", "South", "North", "West", "South", "East", "West", "East", "North", "South"]
}
df = pd.DataFrame(data)
df
```

***2. Basic Exploration***

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

***3. Key Analysis***

```python
# 1. Total Sales
df["Sales"].sum()
```

```python
#2. Sales by Product
df.groupby("Product", as_index=False)["Sales"].sum()
```

```python
#3.Sales by Region
df.groupby("Region", as_index=False)["Sales"].sum()
```

```python
#4. Top Selling Product
df.groupby("Product", as_index=False)["Sales"].sum().sort_values(by="Sales", ascending=False).head(1)
```

```python
#5. Highest Sales Region
df.groupby("Region", as_index=False)["Sales"].sum().sort_values(by="Sales", ascending=False).head(1)
```

***4. Insights***

Insights:
1. Product D has the highest sales.
2. South region contributes high revenue.
3. Some regions have lower sales and need improvement.

***5. Simple Visualization***

```python
import matplotlib.pyplot as plt

df.groupby("Product")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```
